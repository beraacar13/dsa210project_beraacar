Firstly, I merged 22-23, 23-24 Stats and Salary information. Then, I cleaned the players' stats datasets since there were some duplicate elements resulting from player trades. Finally, I saved the merged datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

adv_22 = pd.read_csv('22-23advanced.csv', sep=';', header=1)
adv_23 = pd.read_csv('23-24advanced.csv', sep=';', header=1)
salaries = pd.read_csv('Nba Player Salaries.csv')

def clean_names(name):
    if pd.isna(name): return name
    for suffix in [' III', ' II', ' IV', ' Jr.', ' Sr.']:
        name = name.replace(suffix, '')
    return name.strip()

adv_22['Player_Clean'] = adv_22['Player'].apply(clean_names)
adv_23['Player_Clean'] = adv_23['Player'].apply(clean_names)
salaries['Player_Clean'] = salaries['Player Name'].apply(clean_names)

def get_summary_rows(df):
    return df.sort_values(by=['Player_Clean', 'G'], ascending=[True, False]).drop_duplicates(subset='Player_Clean', keep='first')

clean_adv_22 = get_summary_rows(adv_22)
clean_adv_23 = get_summary_rows(adv_23)

salaries['Salary_22_23'] = salaries['2022/2023'].replace(r'[\$,]', '', regex=True).astype(float)
salaries['Salary_23_24'] = salaries['2023/2024'].replace(r'[\$,]', '', regex=True).astype(float)

merged_22 = pd.merge(clean_adv_22, salaries[['Player_Clean', 'Salary_22_23']], on='Player_Clean')
merged_22['Season'] = '2022-23'
merged_22.rename(columns={'Salary_22_23': 'Salary'}, inplace=True)

merged_23 = pd.merge(clean_adv_23, salaries[['Player_Clean', 'Salary_23_24']], on='Player_Clean')
merged_23['Season'] = '2023-24'
merged_23.rename(columns={'Salary_23_24': 'Salary'}, inplace=True)

final_df = pd.concat([merged_22, merged_23], ignore_index=True)

print(f"Success! A total of {len(final_df)} player/season records were merged.")
final_df.to_csv('NBA_Advanced_Salary_Data.csv', index=False)

```
